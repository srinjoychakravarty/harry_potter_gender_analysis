```python
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

```python
path = r'spell_sheet.csv'

data = pd.read_csv(path)
```

```python
# Create an empty dictionary for words in book
spell_dict = dict()
```

```python
spell_dict = dict(zip(data.Incantation,data.Light))
```

```python
spell_effect = dict(zip(data.Incantation,data.Effect))
```

```python
def getEffectForColor(incantation):
    for key, value in spell_effect.items():
        if key==incantation:
            print(value) 
```

```python
getEffectForColor("Lumos Maxima")
```

```python
def getIncantationForColor(color):
    for key, value in spell_dict.items():
        if value==color:
            print(key) 
```

```python
getIncantationForColor("Blue")
```

```python
spell_type = dict(zip(data.Incantation,data.Type))
```

```python
def getAllIncantationsOfType(type):
    for key, value in spell_type.items():
        if value==type:
            print(key) 
```

```python
getAllIncantationsOfType("Charm")
    
```
